```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
uber_data = pd.read_csv("uber_data.csv")
```

```python
uber_data.head()
```

```python
uber_data.info()
```

```python
uber_data["tpep_pickup_datetime"] = pd.to_datetime(uber_data["tpep_pickup_datetime"])
uber_data["tpep_dropoff_datetime"] = pd.to_datetime(uber_data["tpep_dropoff_datetime"])
```

```python
uber_data.head()
```

```python
uber_data.info()
uber_data.to_csv("uber_data_converted.csv")
```

```python
uber_data[["tpep_pickup_datetime", "tpep_dropoff_datetime"]].drop_duplicates().reset_index(drop=True)

```

```python
uber_data['pick_hour'] = uber_data['tpep_pickup_datetime'].dt.hour
uber_data['pick_day'] = uber_data['tpep_pickup_datetime'].dt.day
uber_data['pick_month'] = uber_data['tpep_pickup_datetime'].dt.month
uber_data['pick_year'] = uber_data['tpep_pickup_datetime'].dt.year
uber_data['pick_weekday'] = uber_data['tpep_pickup_datetime'].dt.weekday
```

```python
uber_data['drop_hour'] = uber_data['tpep_dropoff_datetime'].dt.hour
uber_data['drop_day'] = uber_data['tpep_dropoff_datetime'].dt.day
uber_data['drop_month'] = uber_data['tpep_dropoff_datetime'].dt.month
uber_data['drop_year'] = uber_data['tpep_dropoff_datetime'].dt.year
uber_data['drop_weekday'] = uber_data['tpep_dropoff_datetime'].dt.weekday
```

```python
uber_data[["passenger_count"]].drop_duplicates().reset_index(drop=True)


trip_distance_dim = uber_data[["trip_distance"]].drop_duplicates().reset_index(drop=True)
```

```python
rate_code = {
    1: "Standard Rate",
    2: "JFK",
    3: "Newark",
    4: "Nassau or Westchester",
    5: "Negotiated Fare",
    6: "Group Ride"
}

uber_data[['RatecodeID']].reset_index(drop=True)
uber_data['rate_code_name'] = uber_data['RatecodeID'].map(rate_code)

```

```python
uber_data["dropoff_longitude"]
```

```python
payment_type_name = {
    1:"Credit card",
    2:"Cash",
    3:"No charge",
    4:"Dispute",
    5:"Unknown",
    6:"Voided trip"
}
uber_data[['payment_type']].reset_index(drop=True)
uber_data['payment_type_name'] = uber_data['payment_type'].map(payment_type_name)
```

```python
uber_sample = uber_data.sample(300)
sns.pairplot(uber_sample)
```

```python
uber_data.dtypes
```

```python
from sklearn import preprocessing

label_encoder = preprocessing.LabelEncoder()
obj = (uber_data.dtypes == 'object')
for col in list(obj[obj].index):
    uber_data[col] = label_encoder.fit_transform(uber_data[col])
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(uber_data.corr(), cmap="coolwarm", fmt=".2f",
            linewidths=2, annot=True)
```
